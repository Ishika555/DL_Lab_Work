```python
# CELL 1: SPLIT TOKENIZATION

text = "Deep learning models are transforming AI and NLP"

tokens = text.split()

print("Split Tokens:", tokens)
```

```python
# CELL 2: BPE TOKENIZATION (GPT2)

from transformers import GPT2Tokenizer

text = "Deep learning models are transforming AI and NLP"

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

tokens = tokenizer.tokenize(text)
ids = tokenizer.encode(text)

print("BPE Tokens:", tokens)
print("BPE IDs:", ids)
```

```python
# CELL 3: WORDPIECE TOKENIZATION (BERT)

from transformers import BertTokenizer

text = "Deep learning models are transforming AI and NLP"

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize(text)
ids = tokenizer.encode(text)

print("WordPiece Tokens:", tokens)
print("WordPiece IDs:", ids)
```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

text = "Deep learning models are transforming AI and NLP"

# Save sample text
with open("sample.txt", "w") as f:
    f.write(text)

# Initialize tokenizer
tokenizer = Tokenizer(BPE())
tokenizer.pre_tokenizer = Whitespace()

trainer = BpeTrainer(vocab_size=50)

# Train
tokenizer.train(["sample.txt"], trainer)

# Encode
output = tokenizer.encode(text)

print("Custom BPE Tokens:", output.tokens)
print("Custom BPE IDs:", output.ids)
```

```python
import sentencepiece as spm

# Load trained model
sp = spm.SentencePieceProcessor()
sp.load("m.model")

text = "Deep learning models are transforming AI"

tokens = sp.encode(text, out_type=str)
ids = sp.encode(text, out_type=int)

print("Tokens:", tokens)
print("IDs:", ids)
```

```python
# CELL 1: WordPiece + TensorFlow

from transformers import BertTokenizer
import tensorflow as tf

text = "Deep learning models use embeddings"

# Step 1: WordPiece Tokenization
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text, add_special_tokens=False)

print("Tokens:", tokens)
print("Token IDs:", token_ids)

# Step 2: Embedding
vocab_size = tokenizer.vocab_size
embedding_dim = 8

embedding_layer = tf.keras.layers.Embedding(
    input_dim=vocab_size,
    output_dim=embedding_dim
)

input_tensor = tf.constant([token_ids])
embeddings = embedding_layer(input_tensor)

print("Embeddings shape:", embeddings.shape)
print("Embeddings:\n", embeddings.numpy())
```
